```python
import pandas as pd
df = pd.read_csv('./data/서울시_우리마을가게_상권분석서비스(신_상권_추정매출)_2020년.csv', encoding='cp949')
df
```

```python
df = df[df['상권_코드_명'].str.contains('역')]
df
```

### 상권코드명만 추출

```python
new_df = pd.DataFrame(df['상권_코드_명'].sort_values().unique(), columns=['상권_코드_명'])
new_df
```

### 1. 공백을 구분자로 split
#### 공백을 구분자로 분리하기 전에 특정 값들 변경작업

```python
new_df[(new_df['상권_코드_명'].str.contains(' ')) & (new_df['상권_코드_명'].str.split(' ').str[0].str.contains('역')==False)]
```

```python
new_df.replace({'상권_코드_명':'경의중앙 신촌역'}, {'상권_코드_명':'신촌역'}, inplace=True)
new_df.replace({'상권_코드_명':'서울축산농협 독산역지점(독산파출소)'}, {'상권_코드_명':'독산역'}, inplace=True)
new_df.loc[[27, 272], :]  # 변경됐는지 확인~
```

#### 공백을 구분자로 분리하고, 출구번호 사라지면서 나타난 중복값 제거

```python
new_df['상권_코드_명'] = new_df['상권_코드_명'].str.split(' ').str[0]
new_df.drop_duplicates(inplace=True)
new_df
```

```python
new_df['상권_코드_명'].values
```

### 2. '역'이 포함되어 있는 역이 아닌 행 검색하고 삭제~
#### 조금 노가다. 다른 방식 나중에 찾아보기

```python
del_rows = new_df[(new_df['상권_코드_명'] == '역촌중앙시장')|(new_df['상권_코드_명'] == '강남터미널지하도상점가2구역')\
       |(new_df['상권_코드_명'] == '구역삼세무서')|(new_df['상권_코드_명'] == '역삼월드메르디앙아파트')\
       |(new_df['상권_코드_명'] == '역촌노인복지관')].index
del_rows
```

```python
new_df.drop(del_rows, axis=0, inplace=True)
new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
# 값 확인
new_df['상권_코드_명'].values
```
