```python
import pandas as pd
sell_df = pd.read_csv('./2020_매출데이터_전처리완.csv', encoding='cp949')
sell_df.head()
```

### 동 삭제(그룹화할 때 문제 생김)

```python
sell_df.drop(columns='동', inplace=True)
```

```python
sell_df.columns
```

```python
subway_df = pd.read_csv('./2020_지하철역하차인원_연도_월_일_분기_요일별_매칭완(모든업종).csv', encoding='cp949')
subway_df.drop(['Unnamed: 0'], axis = 1, inplace = True)
subway_df
```

```python
subway_df.columns
```

```python
df_grouped = subway_df.groupby(['역명', '요일'])[['00-11', '11-17', '17-24']].sum()
df_grouped
```

### 지하철데이터 - 요일을 컬럼명으로 (예: 월_00-11시)

```python
pivot_df = subway_df.pivot_table(index=['역명', '분기'], columns='요일', values=['00-11', '11-17', '17-24'], aggfunc='sum', fill_value=0)
pivot_df
```

```python
# 요일 순서를 지정
day_order = ['월', '화', '수', '목', '금']

# 열 순서를 조정하고 컬럼 이름을 변경
pivot_df = pivot_df.reindex(columns=day_order, level=1)
pivot_df.columns = [f'{col[0]}_{col[1]}' for col in pivot_df.columns]
```

```python
pivot_df.columns
```

```python
pivot_df.reset_index(inplace=True)
pivot_df
```

### 분기별로 나눠서 데이터 저장

```python
sell_df.rename(columns={'상권_코드_명':'역명', '기준_분기_코드':'분기', '기준_년_코드':'년'}, inplace=True)
sell_df.columns
```

```python
sell_q1 = sell_df[sell_df['분기'] == 1]
sell_q1
```

```python
subway_q1 = pivot_df[pivot_df['분기'] == 1]
subway_q1
```

```python
# pd.options.display.max_rows = 60
# pd.options.display.max_columns = 20
# row 생략 없이 출력
pd.set_option('display.max_rows', None)
# col 생략 없이 출력
pd.set_option('display.max_columns', None)
```

```python
merge_q1 = pd.merge(sell_q1, subway_q1, on=['역명', '분기'], how='outer')
merge_q1
```

```python
merge_q1.dropna(inplace=True)
merge_q1.reset_index(drop=True, inplace=True)
merge_q1
```

```python
# 2분기
sell_q2 = sell_df[sell_df['분기'] == 2]
subway_q2 = pivot_df[pivot_df['분기'] == 2]
merge_q2 = pd.merge(sell_q2, subway_q2, on=['역명', '분기'], how='outer')
merge_q2.dropna(inplace=True)
merge_q2.reset_index(drop=True, inplace=True)
merge_q2
```

```python
# 3분기
sell_q3 = sell_df[sell_df['분기'] == 3]
subway_q3 = pivot_df[pivot_df['분기'] == 3]
merge_q3 = pd.merge(sell_q3, subway_q3, on=['역명', '분기'], how='outer')
merge_q3.dropna(inplace=True)
merge_q3.reset_index(drop=True, inplace=True)
merge_q3
```

```python
# 4분기
sell_q4 = sell_df[sell_df['분기'] == 4]
subway_q4 = pivot_df[pivot_df['분기'] == 4]
merge_q4 = pd.merge(sell_q4, subway_q4, on=['역명', '분기'], how='outer')
merge_q4.dropna(inplace=True)
merge_q4.reset_index(drop=True, inplace=True)
merge_q4
```

```python
merge_q1.to_csv('./20년 1분기 매출_유동인구_merge.csv', encoding='cp949', index=False)
merge_q2.to_csv('./20년 2분기 매출_유동인구_merge.csv', encoding='cp949', index=False)
merge_q3.to_csv('./20년 3분기 매출_유동인구_merge.csv', encoding='cp949', index=False)
merge_q4.to_csv('./20년 4분기 매출_유동인구_merge.csv', encoding='cp949', index=False)
```
