## ~2020 지하철 9호선(언주-) 데이터 정리~

```python
df20_9 = pd.read_csv('./data/서울교통공사_9호선2-3단계역별일별시간대별승하차인원_20201231.csv', encoding='cp949')
df20_9.head(5)
```

```python
df20_9.columns
```

```python
df20_9['00-11']=df20_9[['00~01', '01~02', '02~03', '03~04','04~05', '05~06', '06~07', '07~08', '08~09', '09~10', '10~11']].sum(axis=1)
df20_9.drop(['00~01', '01~02', '02~03', '03~04','04~05', '05~06', '06~07', '07~08', '08~09', '09~10', '10~11'], axis=1, inplace=True)
```

```python
df20_9.head(5)
```

```python
df20_9['11-17']=df20_9[['11~12','12~13', '13~14', '14~15', '15~16', '16~17']].sum(axis=1)
df20_9.drop(['11~12','12~13', '13~14', '14~15', '15~16', '16~17'], axis=1, inplace=True)
df20_9.head(5)
```

```python
df20_9['17-24']=df20_9[['17~18', '18~19', '19~20','20~21', '21~22', '22~23', '23~24']].sum(axis=1)
df20_9.drop(['17~18', '18~19', '19~20', '20~21', '21~22', '22~23', '23~24'], axis=1, inplace=True)
df20_9.head(5)
```

```python
df20_9 = df20_9[df20_9['구분'] == '순하차']
df20_9.head()
```

### ~1-8호선, 9호선 일부 역 합치기~

```python
df20_subway = pd.concat([df20_1_8, df20_9])
df20_subway.info()
```
