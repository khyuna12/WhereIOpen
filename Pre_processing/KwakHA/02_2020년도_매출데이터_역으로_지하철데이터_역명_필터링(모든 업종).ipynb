### 매출데이터['상권_코드_명'] --> 지하철데이터['역명']

```python
import pandas as pd
df = pd.read_csv('./2020_매출데이터_상권코드명_지하철역_모든업종.csv', encoding='cp949')
df
```

```python
df[df['상권_코드_명'] == '역삼']
```

```python
df2 = pd.read_csv('./2020_지하철역하차인원_연도_월_일_분기_요일별.csv', encoding='cp949')
df2.head()
```

#### 1) 혼자 '역' 붙어 있는 서울역에서 역 제거

```python
df2.replace({'역명':'서울역'}, {'역명':'서울'}, inplace=True)
df2
```

#### 2) 필터링하기 위해 괄호 붙은 거 제거

```python
df2['역명'] = df2['역명'].str.split('\(').str[0]
df2
```

#### 필터링하기 전에 변수로 지정해서 필터링한 후 결과랑 비교

```python
before_filter = df2['역명'].sort_values().unique()
before_filter
```

```python
df2['역명'].nunique()
```

### 매출데이터의 '상권_코드_명' 컬럼으로 지하철데이터 필터링

```python
df2 = df2[df2['역명'].isin(df['상권_코드_명'])]
df2
```

```python
after_filter = df2['역명'].sort_values().unique()
after_filter
```

```python
df2['역명'].nunique()
```

```python
before_filter_df = pd.DataFrame(before_filter, columns=['역명'])
before_filter_df
```

```python
after_filter_df = pd.DataFrame(after_filter, columns=['역명'])
after_filter_df
```

#### 지하철데이터에서 버려진 것들

```python
before_filter_df[before_filter_df['역명'].isin(after_filter_df['역명'])==False]
```
