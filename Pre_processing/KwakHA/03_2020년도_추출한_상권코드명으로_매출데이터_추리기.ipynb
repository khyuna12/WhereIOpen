### 그룹화해서 더하기 전에 출구 구분 없애기
#### 역으로 분리하기 전에 따로 빼둘 것들

```python
separate = df[(df['상권_코드_명'] == '동대문역사문화공원역') | (df['상권_코드_명'] == '역삼역') | (df['상권_코드_명'] == '역촌역')]
separate
```

```python
df['상권_코드_명'] = df['상권_코드_명'].str.split('역').str[0]
df
```

```python
df['상권_코드_명'] = df['상권_코드_명'].str.split('\(').str[-1]
df
```

```python
del_index = df[df['상권_코드_명'] == ''].index
df.drop(del_index, inplace=True)
df
```

```python
df.replace({'상권_코드_명':'디지털미디어시티)'}, {'상권_코드_명':'디지털미디어시티'}, inplace=True)
df
```

```python
df['상권_코드_명'].sort_values().unique()
```

```python
df = pd.concat([df, separate])
df
```

### groupby함수 적용해서 같은 행끼리 더하기

```python
group_df = df.groupby(['상권_코드_명', '자치구', '동', '기준_분기_코드'])['분기당_매출_금액', '분기당_매출_건수','주중_매출_금액', '주말_매출_금액', 
                                               '월요일_매출_금액', '화요일_매출_금액', '수요일_매출_금액', '목요일_매출_금액', 
                                               '금요일_매출_금액', '토요일_매출_금액', '일요일_매출_금액', '시간대_00~06_매출_금액', 
                                               '시간대_06~11_매출_금액', '시간대_11~14_매출_금액', '시간대_14~17_매출_금액', 
                                               '시간대_17~21_매출_금액', '시간대_21~24_매출_금액', '주중_매출_건수', '주말_매출_건수', 
                                               '월요일_매출_건수', '화요일_매출_건수', '수요일_매출_건수', '목요일_매출_건수', 
                                               '금요일_매출_건수', '토요일_매출_건수', '일요일_매출_건수', '시간대_건수~06_매출_건수', 
                                               '시간대_건수~11_매출_건수', '시간대_건수~14_매출_건수', '시간대_건수~17_매출_건수', 
                                               '시간대_건수~21_매출_건수', '시간대_건수~24_매출_건수', '점포수'].sum()
group_df
```

```python
group_df.reset_index(level=['상권_코드_명', '자치구', '동', '기준_분기_코드'], inplace = True)  # 인덱스를 열로 변환
group_df
```

```python
group_df.insert(4, '기준_년_코드', '2020')
group_df
```

```python
group_df['시간대_00~11_매출_금액'] = group_df['시간대_00~06_매출_금액'] + group_df['시간대_06~11_매출_금액']
group_df['시간대_11~17_매출_금액'] = group_df['시간대_11~14_매출_금액'] + group_df['시간대_14~17_매출_금액']
group_df['시간대_17~24_매출_금액'] = group_df['시간대_17~21_매출_금액'] + group_df['시간대_21~24_매출_금액']
group_df.drop(['시간대_00~06_매출_금액', '시간대_06~11_매출_금액', '시간대_11~14_매출_금액', '시간대_14~17_매출_금액',
               '시간대_17~21_매출_금액', '시간대_21~24_매출_금액'], axis=1, inplace=True)

group_df['시간대_00~11_매출_건수'] = group_df['시간대_건수~06_매출_건수'] + group_df['시간대_건수~11_매출_건수']
group_df['시간대_11~17_매출_건수'] = group_df['시간대_건수~14_매출_건수'] + group_df['시간대_건수~17_매출_건수']
group_df['시간대_17~24_매출_건수'] = group_df['시간대_건수~21_매출_건수'] + group_df['시간대_건수~24_매출_건수']
group_df.drop(['시간대_건수~06_매출_건수', '시간대_건수~11_매출_건수', '시간대_건수~14_매출_건수', '시간대_건수~17_매출_건수', 
               '시간대_건수~21_매출_건수', '시간대_건수~24_매출_건수'], axis=1, inplace=True)
group_df
```
